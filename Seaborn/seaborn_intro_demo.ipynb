```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
diamonds = pd.read_csv('../data/Diamonds Prices2022.csv')
```

```python
sns.barplot(x = 'cut', y = 'price', data=diamonds)
```

```python
sns.barplot(x = 'price', y = 'cut', ci = None, data=diamonds)
```

```python
sns.barplot(
    x = 'price',
    y = 'cut',
    data = diamonds,
    facecolor = 'white',
    edgecolor = 'black'
)
```

```python
sns.barplot(
    y = 'cut',
    x = 'price',
    hue = 'clarity',
    data = diamonds,
    palette = 'husl'
)

```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize = (8, 8))

sns.barplot(
    y = 'cut',
    x = 'price',
    hue = 'clarity',
    data = diamonds,
    palette = 'husl'
)

ax.legend(bbox_to_anchor = (1.2, 1))
```

## Histograms

```python
sns.histplot(
    x = 'price',
    data = diamonds.query('clarity in ["I1", "IF"]')
)
```

```python
sns.histplot(
    x = 'price',
    hue = 'clarity',
    data = diamonds.query('clarity in ["I1", "IF"]')
)
```

```python
sns.histplot(
    x = 'price',
    hue = 'clarity',
    bins = 10,
    data = diamonds.query('clarity in ["I1", "IF"]')
)
```

```python
sns.histplot(
    x = 'price',
    hue = 'clarity',
    binwidth = 1000,
    data = diamonds.query('clarity in ["I1", "IF"]')
)
```

```python
sns.histplot(
    x = 'price',
    hue = 'clarity',
    binwidth = 1000,
    stat='percent',
    data = diamonds.query('clarity in ["I1", "IF"]')
)
```

```python
sns.histplot(
    x = 'price',
    hue = 'clarity',
    binwidth = 1000,
    stat='percent',
    palette='husl',
    data = diamonds.query('clarity in ["I1", "IF"]')
)
```

```python
sns.histplot(
    x = 'price',
    hue = 'clarity',
    binwidth = 1000,
    stat='percent',
    palette='husl',
    kde = True,
    data = diamonds.query('clarity in ["I1", "IF"]')
)
```

```python
sns.histplot(
    x = 'price',
    data = diamonds
)
```

```python
sns.histplot(
    x = 'price',
    y = 'clarity',
    data = diamonds
)
```

```python
sns.histplot(
    x = 'price',
    y = 'clarity',
    cbar = True,
    data = diamonds
)
```

## Scatter plots

```python
fig, ax = plt.subplots(figsize = (10, 6))

sns.scatterplot(
    x = 'x',
    y = 'price',
    data = diamonds,
    hue = 'carat',
    palette='husl',
    size = 'carat'
)

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10, 6))

sns.scatterplot(
    x = 'x',
    y = 'price',
    data = diamonds,
    hue = 'clarity',
    palette='viridis',
    edgecolor = None,
    alpha = 0.1
)

plt.show()
```

## Boxplots and violins

```python
fig, ax = plt.subplots(figsize = (10, 6))

sns.boxplot(x = 'price', data=diamonds)

sns.despine()

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8, 5))

sns.boxplot(x = 'cut', y = 'price', data=diamonds)

sns.despine()

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8, 5))

sns.boxplot(y = 'clarity', x = 'price', data=diamonds, color = 'orange')

sns.despine()

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8, 5))

sns.violinplot(x = 'cut', y = 'price', data = diamonds)

sns.despine()

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8, 8))

sns.violinplot(y = 'clarity', x = 'price', data = diamonds, palette='viridis')

sns.despine()

plt.show()
```
