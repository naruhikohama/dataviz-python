```python
import pandas as pd
import seaborn as sns
```

```python
diamonds = pd.read_csv('../data/Diamonds Prices2022.csv')
```

```python
sns.barplot(x = 'cut', y = 'price', data=diamonds)
```

```python
sns.barplot(x = 'price', y = 'cut', ci = None, data=diamonds)
```

```python
sns.barplot(
    x = 'price',
    y = 'cut',
    data = diamonds,
    facecolor = 'white',
    edgecolor = 'black'
)
```

```python
sns.barplot(
    y = 'cut',
    x = 'price',
    hue = 'clarity',
    data = diamonds,
    palette = 'husl'
)

```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize = (8, 8))

sns.barplot(
    y = 'cut',
    x = 'price',
    hue = 'clarity',
    data = diamonds,
    palette = 'husl'
)

ax.legend(bbox_to_anchor = (1.2, 1))
```
