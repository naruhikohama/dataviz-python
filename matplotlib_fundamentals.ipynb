```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import datetime
```

```python
housing_raw = pd.read_csv('data/housing_data.csv',
                          parse_dates=['period_begin', 'period_end'],
                          dtype={'total_homes_sold':'Int32', 'region_name':'category'})
```

```python
housing_raw.head()
```

```python
housing_raw.info()
```

```python
ca_housing = (housing_raw
              .loc[(housing_raw['region_name'].str.contains('CA'))]
              .assign(region_name = housing_raw['region_name'].str[0:-11]))
```

```python
ca_housing.info(memory_usage = 'deep')
```

```python
ca_housing.head()
```

```python
ca_housing_pivot = ca_housing.pivot_table(index='period_begin',
                                          columns='region_name',
                                          values='median_active_list_price').assign(
                                            CA_average = lambda x: x.mean(axis = 1)
                                          )
```

```python
ca_housing_pivot.head()
```

```python
ca_housing_markets = ca_housing_pivot.loc[:, ['San Francisco',
                                              'Los Angeles',
                                              'San Diego',
                                              'CA_average']]
ca_housing_markets                                              
```

```python
fig, ax = plt.subplots()

fig.set_figheight(6)
fig.set_figwidth(10)

# ax.plot(ca_housing_markets)

ax.plot(ca_housing_markets.index, ca_housing_markets['San Francisco'], label = 'SF')
ax.plot(ca_housing_markets.index, ca_housing_markets['Los Angeles'], label = 'LA')

fig.suptitle("House Prices", fontsize = 16)
ax.set_title('CA Markets 2017 - 2022', fontsize=14)
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Price (Millions)', fontsize=12)

ax.legend(bbox_to_anchor = (.6, -.1), ncol = 2, frameon = False)

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10, 7))

ax.plot(
    ca_housing_markets.index,
    ca_housing_markets['San Francisco'] / 1_000_000,
    label = 'San Francisco',
    color = 'green',
)

ax.plot(
    ca_housing_markets.index,
    ca_housing_markets['Los Angeles'] / 1_000_000,
    label = 'Los Angeles',
    color = 'purple',
)

ax.set_title('Housing prices in CA Markets 2017-2022', fontsize = 12)
ax.set_xlabel("Year")
ax.set_ylabel("Price (Millions)")
ax.set_ylim(0, 1.6)

line_max_y = ca_housing_markets[ca_housing_markets.index == datetime.datetime(2018, 3, 5)][['San Francisco', 'Los Angeles']].values.max() / 1_000_000

ax.axvline(datetime.datetime(2018, 3, 5), ymin= 0, ymax = line_max_y / 1.6, color = 'grey', ls = 'dashed', label = 'random date')

ax.legend()

plt.show()
```
