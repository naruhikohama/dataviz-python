```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = [x for x in range(10)]
data2 = [x*2 for x in range(10)]
```

```python
fig, ax = plt.subplots()

ax.plot(
    data,
    c = 'green'
)

ax.plot(
    data2,
    c = 'blue'
)

plt.show()
```

```python
plt.rcParams['axes.prop_cycle'] = plt.cycler(color = ['red', 'orange', 'grey', 'green'])
```

```python
fig, ax = plt.subplots()

ax.plot(
    data,
)

ax.plot(
    data2,
)

plt.show()
```

```python
plt.rcParams['axes.prop_cycle'] = plt.cycler(color = plt.cm.Dark2.colors)
```

```python
fig, ax = plt.subplots()

ax.plot(
    data,
)

ax.plot(
    data2,
)

plt.show()
```
