```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
diamonds = (
    pd.read_csv('data/Diamonds Prices2022.csv')
    .drop(['Unnamed: 0'], axis = 1)
)
```

```python
diamonds.head()
```

```python
fig, ax = plt.subplots()

ax.scatter(
    diamonds[diamonds['carat'] < 2.25]['carat'],
    diamonds[diamonds['carat'] < 2.25]['price'],
    alpha=0.3
)

ax.set_title('Price per carat')
ax.set_ylabel('Price in US$')
ax.set_xlabel('carat')

plt.show()
```

# Histograms

```python
fig, ax = plt.subplots()

ax.hist(
    diamonds['carat'],
    bins = 20,
    density = True,
    edgecolor = 'white',
    lw = 2
)

plt.show()
```

```python
fig, ax = plt.subplots()

ax.hist(
    diamonds[diamonds['cut'] == 'Ideal']['carat'],
    bins=20,
    edgecolor='white',
    alpha = 0.5,
    density = True
)

ax.hist(
    diamonds[diamonds['cut'] == 'Fair']['carat'],
    bins = 20,
    edgecolor = 'white',
    alpha = 0.5,
    density = True
)

ax.legend(['Ideal', 'Fair'])

plt.show()
```
